```python
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=200,
                n_features=2,
                centers=2,
                random_state=2)
```

```python
print(X.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],
           c='gray',
           edgecolor='black',
           marker='o')
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=2,
                     random_state=0)
gmm.fit(X)

label_gmm = gmm.predict(X)
print(label_gmm)
```

```python
import pandas as pd
gmm_columns = ['gmm_comp1','gmm_comp2']
X_gmm_df = pd.DataFrame(X,columns=gmm_columns)
X_gmm_df['target']=y
X_gmm_df['label_gmm'] = label_gmm
X_gmm_df.head(5)
```

```python
print(set(X_gmm_df['target']))
print(set(X_gmm_df['label_gmm']))
```

```python
df = X_gmm_df
markers=['o','x']

for i,mark in enumerate(markers):
    df_i = df[df['label_gmm']==i]
    target_i =i
    X1 = df_i['gmm_comp1']
    X2 = df_i['gmm_comp2']    
    plt.scatter(X1,X2,marker=mark,
               label=target_i)
    
plt.xlabel('gmm_component1')
plt.ylabel('gmm_component2')
plt.legend()
plt.show()
```

```python
df = X_gmm_df
markers=['o','x']

for i,mark in enumerate(markers):
    df_i = df[df['target']==i]
    target_i =i
    X1 = df_i['gmm_comp1']
    X2 = df_i['gmm_comp2']    
    plt.scatter(X1,X2,marker=mark,
               label=target_i)
    
plt.xlabel('gmm_component1')
plt.ylabel('gmm_component2')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sil_score = silhouette_score(X, label_gmm)
print(sil_score)
```
