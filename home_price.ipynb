```python
import numpy as np
#수학적 계산 관련 함수 제공
import pandas as pd
#데이터를 데이터 프레임 형식으로 다루는 라이브러리. 데이터 이용시 필수
import seaborn as sns
#시각화 1 더 화려함
import matplotlib.pyplot as plt
#시각화 2 덜 화려함
from sklearn import datasets
#사이킷런에서 제공하는 다양한 데이터를 이용 가능
```

```python
from sklearn.datasets import fetch_openml

raw_boston = fetch_openml(name='boston')
#이게 원래는 raw_boston = datasets.load-boston()으로 해야하는데 사이킷런 1.2 이후로 삭제 되서 이거로 대체함
X_boston = pd.DataFrame(raw_boston.data)
#data = feature = 입력 데이터 부분
Y_boston = pd.DataFrame(raw_boston.target)
#target = 출력 데이터
df_boston = pd.concat([X_boston,Y_boston],axis=1)
#입출력 데이터를 결합하여 전체 데이터 프레임 제작
##axis=0이면 위+아래 형태로 , axis=1이면 좌+우 형태로 합침
###피처 와 타깃은 보통 좌우로 합친다
```

```python
len(df_boston)
#len은 길이를 반환하는 함수인데, 데이터 프레임의 전체 row를 반환해줌
##즉, 506개의 데이터로 이뤄짐. 집이 506채라는 뜻 
```

```python
df_boston.head()
#무조건 0~4행을 보여주는 함수임. 물론 (n)으로 출력 개수 커스텀 가능
##원래 데이터는 이름이 없고 숫자만 반환하는데 이는 이렇게 설정해야함
feature_boston = raw_boston.feature_names
##이건 보스턴 데이터의 피처 이름을 확인하는 거고
#print(feature_boston)
##이걸로 피처들의 이름만 뜸
col_boston=np.append(feature_boston,['target'])
#원하는건 피처 + 타겟 형태이므로 append함수를 통해 연결해줌
df_boston.columns = col_boston
df_boston.head()
```
